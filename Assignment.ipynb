```
#Import nsepy and plot the stock from 1 jan 2015 to 30 Dec 2015 (one year total)
from nsepy import get_history
from datetime import date
data = get_history(symbol="SBIN", start=date(2015,1,1), end=date(2015,12,30))
data[['Close']].plot()
```

```
#get data froms sbin for one year
from datetime import date
from nsepy import get_history
sbin = get_history(symbol='SBIN',
                   start=date(2015,1,1),
                   end=date(2015,12,30))
print(sbin)
```

Stock futures price history



```
#create rolling window of size 10.
sbin.rolling(window=10).mean()
```

```
#import pandas and numpy
import numpy as np
import pandas as pd
```

```
#rolling window of size 10 and store in short_rolling
short_rolling = sbin.rolling(window=10).mean()

#visualize for 1 year with matplotlib
short_rolling[['Close']].plot()
```

```
#rolling window size of 75.
short_rolling = sbin.rolling(window=75).mean()

#visualize for 1 year with matplotlib
short_rolling[['Close']].plot()
```
